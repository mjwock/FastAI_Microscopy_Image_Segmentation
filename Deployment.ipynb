```
%reload_ext autoreload
%autoreload 2
%matplotlib inline 
%tensorflow_version 1.x

import os
import pandas as pd
import matplotlib.pyplot as plt

from fastai.callbacks import SaveModelCallback


from torch import nn
from torchsummary import summary
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.metrics import confusion_matrix, matthews_corrcoef, fbeta_score, jaccard_score
import warnings

from datetime import datetime
from ipywidgets import widgets

from deepflash import unetadaption, utility
from deepflash.fastai_extension import *
```

# LOAD DATA

```
PROJECT_FOLDER = '/content/drive/My Drive/FastAI_Microscopy_Segmentation/data' 
```

```
LOAD_MODEL = f'{PROJECT_FOLDER}/model/lambda25_Weights7_200903_0929/final_model.pkl'

TILE_SHAPE = (540,540)
MASK_SHAPE = (356,356)
```

```
IMAGE_DIR = f'{PROJECT_FOLDER}/images/LabWue/red'
SAVE_RESULTS_TO = f'{PROJECT_FOLDER}/predictions/LabWue'
```

LOAD REQUESITES

```
images, _, names = utility.load_images_and_labels(IMAGE_DIR,IMAGE_DIR)
```

# PREDICT

```
learn = load_learner(LOAD_MODEL,'')
```

```
# initiate TileGenerator instance
tiler = TileGenerator(images,learn, TILE_SHAPE, MASK_SHAPE)
```

```
# predict all
tiler.predict_all()
```

```
results, _ = tiler.stitch_results()
```

# SAVE FILES

```
save_to = SAVE_RESULTS_TO.strip('/')
if not os.path.exists(save_to):
  os.makedirs(save_to)
```

```
for image, name in zip(results,names):
  image.save(f'//{save_to}/{name}.png')
```
